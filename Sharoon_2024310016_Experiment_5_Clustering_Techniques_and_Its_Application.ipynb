```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
```

```python
# Generate synthetic data
n_samples = 1000
X, y = make_blobs(n_samples=n_samples, centers=5, cluster_std=0.60, random_state=0)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title('Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()
```

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=5, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Plot K-Means Results
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()

# Silhouette Score
silhouette_kmeans = silhouette_score(X, y_kmeans)
print(f'Silhouette Score for K-Means: {silhouette_kmeans:.2f}')
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X_scaled)

# Plot DBSCAN Results
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=30, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()

# Silhouette Score
if len(set(y_dbscan)) > 1:  # At least 2 clusters are needed to calculate the score
    silhouette_dbscan = silhouette_score(X_scaled, y_dbscan)
    print(f'Silhouette Score for DBSCAN: {silhouette_dbscan:.2f}')
else:
    print("DBSCAN found only one cluster or noise.")
```

```python
# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot PCA Results
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.title('PCA Visualization of K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()
```

```python
# t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=0)
X_tsne = tsne.fit_transform(X)

# Plot t-SNE Results
plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.title('t-SNE Visualization of K-Means Clustering')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.grid()
plt.show()
```
