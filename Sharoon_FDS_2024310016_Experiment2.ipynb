```python
sns.histplot(penguins['bill_length_mm'], kde=True)
plt.show()
```

```python
sns.countplot(x='class', data=titanic)
plt.show()
```

```python
sns.countplot(x='species', data=penguins)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.boxplot(x='class', y='fare', data=titanic)
plt.show()
```

```python
sns.pairplot(penguins)
plt.show()
```

```python
# Convert non-numeric columns to numeric representation before calculating correlations.
tips['sex'] = tips['sex'].map({'Female': 0, 'Male': 1})
tips['smoker'] = tips['smoker'].map({'No': 0, 'Yes': 1})
tips['day'] = tips['day'].map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})
tips['time'] = tips['time'].map({'Lunch': 0, 'Dinner': 1})
sns.heatmap(tips.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
plt.title('Fare Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
```
